## [6] 추론 & 결과 확인 *(수정 금지 — 스크린샷 캡처)*

```python
# ===== 수정 금지 =====
print(f"\n학번: {STUDENT_ID}  이름: {STUDENT_NAME}")
# ===== 수정 금지 끝 =====

# ===== 수정 금지 =====
model.eval()
print("\n[추론 결과]")
with torch.no_grad():
    total, positive      = 0, 0
    total_text, pos_text = 0, 0

    for i in range(n_four_words):
        string  = four_words[i]
        one_hot = torch.from_numpy(word_to_onehot(string)).type_as(torch.FloatTensor())
        hidden  = model.init_hidden()

        input  = one_hot[0:-1]
        input  = torch.unsqueeze(input, 1)
        output, hidden = model(input, hidden)
        output = output.squeeze()   # [3, 26]

        output_string = string[0]
        for j in range(output.size()[0]):
            output_string += onehot_to_word(output[j].data)
            total_text += 1
            if string[j+1] == output_string[-1]:
                pos_text += 1

        total += 1
        if string[-1] == output_string[-1]:
            positive += 1
        print(f'{i+1:2d}  GT: {string}   OUT: {output_string}')

    print(f'\n마지막 글자 정확도: {positive}/{total} ({positive/total:.4f})')
    print(f'전체 글자 정확도   : {pos_text}/{total_text} ({pos_text/total_text:.4f})')
# ===== 수정 금지 끝 =====
```
